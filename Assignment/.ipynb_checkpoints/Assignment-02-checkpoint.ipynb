### 1. 获取地铁路线-站点东dict

````python
def get_metro_map(url, metro_map = {}):
    """
    Return {metro_lin:{station,...}, ...}
    param: url
    return: {str: {str, ...}, ...}
    """
    station = []
    response = requests.get(url)
    html = response.text
    result = json.loads(html)
    for i in result['l']:  # each subway line
        station = [line['n'] for line in i['st']]
        metro_map[i['ln']] = station
    return metro_map
````

### 2. 根据站点请求高德地图api获取地铁站点经纬度

````python
metro_a = "八通线"
station_a = "四惠"
addr = f"北京地铁{metro_a}{station_a}地铁站"
print(addr)
````

````python
def get_station_geoinfo(address):
    """
    Return the latitude and longitude of address.
    param: address: str
    return:  float(latitude), float(longitude)
    """
    key = '0a14cc54e12b1d078fd45d07932564a6'
#     print(address)
    url = f"https://restapi.amap.com/v3/geocode/geo?city=beijing&output=json&address={address}&key={key}"
#     print(url)
    response = requests.get(url).text
#     print(response)
#     print("="*50)
    result = json.loads(response)
    geo = ''
    if len(result['geocodes']) > 0:
        geo = result['geocodes'][0]['location']
    else:
        pass
    
    if len(geo.split(',')) > 1:
        lng, lat = geo.split(',')
        return tuple([float(lat), float(lng)])
    else:
        return tuple('')

def get_single_line_geoinfo(metro_line, stations):
    """
    Return dictionary with format of {station: (lng, lat), ...}
    param: metro_line: str, stations: list
    return: dict
    referce：https://lbs.amap.com/api/webservice/guide/api/georegeo
    """
    station_geo = {}
    
    for station in stations:

        address = f"北京地铁{metro_line}{station}地铁站"
        station_geo[station] = get_station_geoinfo(address)
#         print(station_geo[station])
    return station_geo

def get_all_line_geoinfo(metro_map, metro_line_geoinfo = {}):
    """
    Return all metro lines geoinfo with format {metro_line: {station, ...}, ...}
    param: metro_map: dict
    return: dict
    """
    
    for metro_line, stations in metro_map.items():
        
        line_info = get_single_line_geoinfo(metro_line, metro_map[metro_line])  # {line, {station, ...}}
    
        print(line_info)
        metro_line_geoinfo[metro_line] = line_info
        
#         print(metro_line)
    return metro_line_geoinfo
````

````python
def zip_station(metro_line):
    """
    Return a list of previous station and next station combined. [(a, b), (b, c). ...]
    param: metro_line: str
    return: list
    """
    station1 = list(metro_line.keys())
    station2 = list(metro_line.keys())
    station2.pop(0)
    station_connection = list(zip(station1, station2))
    return station_connection
````

````python
# 查询A点到B点的路径，[A, [B, C]]
# 1个节点存待遍历的节点
# 1个list存遍历过的节点，如果遍历过则直接跳过
# 
def search(start, destination, station_graph):
    
    pathes =[[start]]
    visited = set()  # 遍历过节点
    while pathes:    # if we find existing pathes
        path = pathes.pop(0)
        frontier = path[-1]    # 类似于先进后出，dfs，
        
        if frontier in visited: continue
            
#         successors = connection_graph[froninter]    # 返回后续路径点list
        
        for city in station_graph.neighbors(frontier):
            if city in path:continue    # eliminate loop
            
            new_path = path + [city]
            pathes.append(new_path)
            
            if city == destination:
                target_list.append(new_path)
                pathes.pop()
        visited.add(frontier)
        return pathes
#     if target_list:
#         target_list = sort_candidate(target_list)[0]
#         return ' -> '.join(target_list)
#     else:
#         return f"No way from {start} to {destination}."
````
